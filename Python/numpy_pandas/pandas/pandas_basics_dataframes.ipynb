```python
# read james csv ....create idex with first column now apply functionality in such a way if year is from 1960 to 1975 ...we should get output "Hit"  ...else if 1975 to 2000 flop and any else average
```

```python
jamesseries=pd.read_csv("C:\\Users\miles\\Documents\\GitHub\\Training_futurense\\Python\\numpy_pandas\\sample datasets\\jamesbond_user.csv",usecols=['Year']).squeeze()
type(jamesseries)
```

```python
def check_year(year):
    try:
        if year is None:
            return ""
        elif year>=1960 and year<=1975:
            return "Hit"
        elif year>=1975 and year<=2000:
            return "Flop"
        else:
            return "Average"
    except Exception as err:
        print(err)
```

```python

jamesseries.map(check_year)
```
